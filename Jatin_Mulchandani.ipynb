# The aim of the case study is to identify the risky loan applicants and hence cutting down the amount of credit loss.

### To do this, we will divide our variables into two types : 
#### -> Acquisition - The variables that are available at the time of acquisition(before giving a loan, at the time of loan request).
#### -> Behavioural - The variables that are available at the time when user has taken a loan but is not past the due date.

#### *To identify risky loan applicants we can only consider acquisition type variables for our analysis since they will be available at the time when a user will request a loan.

```python
# List of useful acquisition variables
acquisition = ['id','loan_amnt','term','int_rate','grade','sub_grade','emp_length','home_ownership','annual_inc',
               'verification_status','purpose','addr_state','dti','issue_d','loan_status']
```

### Variable Understanding/Definition
#### id - Primary key/Identifier
#### loan_amnt - Requested loan amount
#### term - Reuqested loan term (36 months or 60 months)
#### int_rate - The interest rate that will be charged based on factor dependent on the user
#### grade - A level assigned to each user
#### sub_grade - A sub-level assigned to each user
#### emp_length - Length of employment of the user
#### home_ownership - whether the user owns a house or not
#### annual_inc - Annual income of the user
#### verification_status - Different verification status on whether user is KYC verified, Income source verified, etc
#### purpose - Purpose of taking the loan
#### addr_state - The address state of the user
#### dti - debt to income ratio to give a brief understanding of user profile
#### issue_d - Month in which loan is funded
#### loan_status - whether the user has defaulted or not or whether he is in the process after taking a loan(will be used as a target)

```python
df_acq = df[acquisition] 
```

```python
df_acq = df_acq.drop_duplicates()
```

```python
### Checking for missing data
#finding percentage of null or missing values

percent_missing = df_acq.isnull().sum() * 100 / len(df_acq)
missing_value_df = pd.DataFrame({'column_name': df_acq.columns,
                                 'percent_missing': percent_missing})

#printing the columns with more than 0% missing values
missing_value_df[missing_value_df['percent_missing']>0].sort_values('percent_missing', ascending = False)
```

### Since Current category of loan_status contains all the ongoing loans, a target can not be defined for such cases and hence we will only use Fully paid loans which can be classed as non defaulters(0) and Charged Off loans which can be classed as defauters(1).

```python
# Removing Current loans from our analysis
df_acq = df_acq[df_acq['loan_status']!='Current']
```

```python
df_acq.shape
```

```python
df_acq['loan_status'].value_counts()
```

```python
# Converting loan_status variable to target by replacing Fully paid with 0(non defaulted) and Charged Off with 1(defaulted)
df_acq['loan_status'] = np.where(df_acq['loan_status']=='Fully Paid',0,1)
```

```python
df_acq.head()
```

```python
df_acq.term.unique()
```

```python
# removing whitespace from term feature
df_acq['term'] = df_acq.term.str.strip()
```

```python
# Stripping % from int_rate to convert it into float type
df_acq['int_rate'] = df_acq.int_rate.str.strip('%').astype(float)
```

```python
#converting issue_d to datetime format
df_acq['issue_d'] = pd.to_datetime(df_acq.issue_d, format='%b-%y')
```

```python
# Deriving some datetime variables
df_acq['issue_d_year'] = df_acq.issue_d.dt.year
df_acq['issue_d_month'] = df_acq.issue_d.dt.strftime('%b')
df_acq['issue_d_weekday'] = df_acq.issue_d.dt.weekday
#data type conversion of year and weekday
df_acq['issue_d_year'] = df_acq['issue_d_year'].astype(object)
df_acq['issue_d_weekday'] = df_acq['issue_d_weekday'].astype(object)
```

# Univariate and Bivariate Analysis

### --- Categorical Variables

```python
# Listing all the categorical variables
cat_list = list(df_acq.select_dtypes(include=['object']).columns)
```

```python
# Printing the category count for each of the categorical variables
for i in df_acq[cat_list]:
    print(df_acq[i].value_counts())
```

### Plotting the distribution of each categorical variable

```python
for i in df_acq[cat_list]:  
    df_acq[i].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
    plt.title(i+' plot')
    plt.ylabel('Count')
    plt.xlabel(i)
    plt.show()
```

## _Observations_
#### Looking at the frequency distribution, we can tell that most of the people are looking for short term loans(36 months)
#### Frequency distribution is nearly rank ordered in grade and sub grade variables with A(x) having highest users to G(x) having the least
#### Most of the individuals applying for loans have a stable employement(10+ years)
#### Users usually do not own their house
#### Debt consolidation and credit card payment are the two major reasons to apply for a loan
#### There's a high percentage of people applying for loan from California
#### The number of loan applicants increase gradually towards the festive period(CHRISTMAS) and then decreases after the festive period is over

### Default Rate analysis across categorical variables

```python
print('The overall default rate for the entire portfolio is :',round(df_acq['loan_status'].mean()*100,2),'%')
```

```python
# Checking the default percentages across differnt categories to see if there's any difference along with population size to check the significance
for i in df_acq[cat_list]:
    x = df_acq.groupby([i]).agg(default_rate=('loan_status', 'mean'),loan_count=('loan_status', 'count'))
    x['default_rate'] = x['default_rate']*100
    print(x)
```

```python
# Plotting default rates across various categories
for i in df_acq[cat_list]:
    x = df_acq[['loan_status',i]].groupby(i).mean()
    x['loan_status'] = x['loan_status']*100
    print(x.plot(kind='bar', figsize=(10,5)))
    plt.title(i+' default percentages')
    plt.ylabel('default_percentage')
    plt.xlabel(i)
    plt.show()
```

## _Observations_

#### Users with short terms loans tend to default less and are hence better suited to be approved for a loan.
#### The lower the grade(with A being higher and G being lower) the more the default percentage.
#### Counter Intuitively, unverified users have lower default percentages compared to rest.
#### Users taking loan for small business tend to have an extremely high default percentage(~27%) while users taking loan for car, credit card, wedding and major purchases tend to perform better.
#### Loans taken from eastern region of USA(MA, PA, VA, ETC...) tend to perform better in terms of default.
#### Default rates tend to get higher moving towards the festive month of Christmas(May to December) and are less for the intial part of the year(January to April).

## _Observations_
#### We can observe some outliers in loan_amt and int_rates but some extreme outliers in annual_inc so we have to perform some outlier treatment on the data

### Outlier Treatment

```python
# We will only treat annual_inc in this case as it has extreme outliers which has also skewed the entire distribution
Q1 = np.percentile(df_acq['annual_inc'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df_acq['annual_inc'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
print('The interquartile range for annual inc is :',IQR)
```

```python
# Removing the outliers in this case. We can also cap them up to a range if needed.
df_acq = df_acq[df_acq['annual_inc'] < (Q3+1.5*IQR)]
```

```python
#Checking whether outlier treatment has worked

#Increasing the figure size of plot
plt.figure(figsize=(12,4))
#Setting subplot index
plt.subplot(1,2,1)
#Histogram plot
sns.distplot(a=df_acq['annual_inc'], rug=True)
plt.subplot(1,2,2)
#Box plot
sns.boxplot(data=df_acq['annual_inc'])
#Single title for both subplots.
plt.suptitle('annual_inc distribution and box plot')
plt.show()
```

### Let us break the numerical continuous variables into segments at their mean and observe whether there are some difference 

```python
# Creating buckets for each numerical variable
df_acq_num['loan_amnt_bucket'] = np.where(df_acq_num['loan_amnt']>df_acq_num['loan_amnt'].mean(),'loan_amnt_high','loan_amnt_low')
df_acq_num['int_rate_bucket'] = np.where(df_acq_num['int_rate']>df_acq_num['int_rate'].mean(),'int_rate_high','int_rate_low')
df_acq_num['emp_length_bucket'] = np.where(df_acq_num['emp_length']>df_acq_num['emp_length'].mean(),'emp_length_high','emp_length_low')
df_acq_num['annual_inc_bucket'] = np.where(df_acq_num['annual_inc']>df_acq_num['annual_inc'].mean(),'annual_inc_high','annual_inc_low')
df_acq_num['dti_bucket'] = np.where(df_acq_num['dti']>df_acq_num['dti'].mean(),'dti_high','dti_low')
```

```python
# A list of all the bucket columns
bucket_list = ['loan_amnt_bucket', 'int_rate_bucket','emp_length_bucket', 'annual_inc_bucket', 'dti_bucket']
```

```python
# Plotting default rates across various buckets
for i in df_acq_num[bucket_list]:
    x = df_acq_num[['loan_status',i]].groupby(i).mean()
    x['loan_status'] = x['loan_status']*100
    print(x.plot(kind='bar', figsize=(10,5)))
    plt.title(i+' default percentages')
    plt.ylabel('default_percentage')
    plt.xlabel(i)
    plt.show()
```

## _Observations_
#### Users with lower loan amount tend to default less than users with higher loan amount. Giving loans of higher amounts is riskier.
#### Users with lower interest rates tend to perform significantly better and default way less than users with higher interest rates.
#### Employment length has no significant impact on the default rates.
#### Users with higher annual income perform way better than users with low income.
#### Users with higher debt to income ratio tend to default more than users with less debit to income ratio.
